```python
# logica de programação

# passo 0 - entender o desafio que você quer resolver

# passo 1 - percorrer todos os arquivos da pasta de dados (pasta vendas)

# passo 3 - tratar/compilar base dados

# passo 4 - calcular o prroduto mais vendido (em quantidade)

# passo 5 - calcular o produto que mais faturou (em faturamento)

# passo 6 - calcular loja com mais faturamento - criar um grafico
```

```python
#só remover os "#" para ver os comandos funcionando, remove the "#" to see the code working regularly
import os
import pandas as pd
import plotly.express as px

#definindo tabelas, todos os arquivos foram oferecidos pela Hashtag Programação em seu curso gratuito de python no youtube
tabela_total = pd.DataFrame()

lista_arquivo = os.listdir("/content/drive/MyDrive/curso python/Vendas")
for arquivo in lista_arquivo:
  if "Vendas" in arquivo:
    tabela = pd.read_csv(f"/content/drive/MyDrive/curso python/Vendas/{arquivo}")
    tabela_total = tabela_total.append(tabela)

#mostrar tabela com tudo
    #display(tabela_total)

#calculos de indicadores

#loja que mais faturou por quantidade
print ('first table')
tabela_produtos = tabela_total.groupby('Produto').sum()
tabela_produtos = tabela_produtos[["Quantidade Vendida"]].sort_values(by = "Quantidade Vendida", ascending=True)
display(tabela_produtos)

#loja que mais faturou(em faturamento)
tabela_total['Faturamento'] = tabela_total['Quantidade Vendida']* tabela_total['Preco Unitario']

print ('Second table')
tabela_faturamento = tabela_total.groupby('Produto').sum()
tabela_faturamento = tabela_faturamento[["Faturamento"]].sort_values(by = "Faturamento", ascending=True)
display(tabela_faturamento)

#loja/cidade que mais faturou em faturamento
print ('Third table')
tabela_loja = tabela_total.groupby('Loja').sum()
tabela_loja = tabela_loja[['Faturamento']]
display (tabela_loja)

#dashboard
grafico = px.bar(tabela_loja, x=tabela_loja.index, y='Faturamento')
grafico.show()
```
